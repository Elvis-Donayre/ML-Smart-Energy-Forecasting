```python
def calcular_ciclo_diario(timestamps):
    """
    Convierte timestamps a un ciclo diario normalizado (0-1).
    
    Args:
        timestamps: DatetimeIndex
    
    Returns:
        Series con valores entre 0 (medianoche) y 1 (23:59)
    """
    return (timestamps.hour + timestamps.minute / 60) / 24

# Uso
datos_fe['ciclo_diario'] = calcular_ciclo_diario(datos_fe.index)
```

### **Efecto Diario**

El modelo debe reconocer que día de la semana es.

```python
def agregar_dia_espanol(df, nombre_columna='dia'):
    """
    Agrega una columna con el nombre del día en español al DataFrame.
    
    Args:
        df: DataFrame con índice de tipo datetime
        nombre_columna: Nombre de la nueva columna (default: 'dia')
    
    Returns:
        DataFrame con la nueva columna agregada
    """
    days_translation = {
        'Monday': '1Lunes',
        'Tuesday': '2Martes', 
        'Wednesday': '3Miércoles',
        'Thursday': '4Jueves',
        'Friday': '5Viernes',
        'Saturday': '6Sábado',
        'Sunday': '0Domingo'
    }
    
    df[nombre_columna] = df.index.day_name().map(days_translation)
    return df

# Uso
datos_fe = agregar_dia_espanol(datos_fe)
```

```python
def analizar_patrones_demanda_reales(df):
    """Analizar patrones reales de demandas"""
    
    # Calcular demanda promedio por hora del día
    demanda_por_hora = df.groupby(df.index.hour)['Demanda'].agg([
        'mean', 'median', 'std', 'min', 'max'
    ]).round(2)
    
    print("=== ANÁLISIS DE PATRONES DEMANDA ===")
    print(demanda_por_hora)
    
    # Identificar automáticamente los percentiles
    demanda_promedio_global = df['Demanda'].mean()
    percentil_75 = df['Demanda'].quantile(0.75)
    percentil_25 = df['Demanda'].quantile(0.25)
    
    print(f"\nDemanda promedio global: {demanda_promedio_global:.2f}")
    print(f"Percentil 75: {percentil_75:.2f}")
    print(f"Percentil 25: {percentil_25:.2f}")
    
    # Identificar horas de alto/bajo consumo por percentiles
    horas_alto_consumo = demanda_por_hora[demanda_por_hora['mean'] > percentil_75].index.tolist()
    horas_bajo_consumo = demanda_por_hora[demanda_por_hora['mean'] < percentil_25].index.tolist()
    
    print(f"Horas de ALTO consumo (>P75): {horas_alto_consumo}")
    print(f"Horas de BAJO consumo (<P25): {horas_bajo_consumo}")
    
    return demanda_por_hora, horas_alto_consumo, horas_bajo_consumo

# Ejecutar análisis
patrones, horas_pico, horas_valle = analizar_patrones_demanda_reales(datos_fe)
```

Valle de demanda (madrugada): 1-6 AM
Consumo más bajo del día
Rango: ~6,000-6,200 MW
Rampa ascendente: 7-11 AM
Subida gradual hacia el pico
La hora 11 ya es pico alto
Meseta alta: 11 AM - 9 PM
Consumo sostenido alto durante todo el día
Pico absoluto: 7-8 PM (19-20h) con ~7,300 MW
Descenso nocturno: 10 PM - 12 AM
Bajada gradual hacia el valle

```python
def crear_features_demanda_especificos(df):
    """
    Crear features específicos basados en patrones reales de demanda eléctrica
    
    Patrones identificados:
    - Valle nocturno: 1-6 AM (consumo <P25)
    - Pico alto: 11 AM, 7-8 PM (consumo >P75) 
    - Rampa ascendente: 7-11 AM
    - Meseta alta: 11 AM - 9 PM
    - Descenso nocturno: 10 PM - 12 AM
    """
    df = df.copy()  # ← Cambio aquí: usar el parámetro df en lugar de datos_fe
    
    # Asegurar que el índice sea datetime
    if not isinstance(df.index, pd.DatetimeIndex):
        df.index = pd.to_datetime(df.index)
    
    # === PATRONES IDENTIFICADOS EN TUS DATOS ===
    
    # Valle nocturno (horas 1-6: consumo <P25)
    df['es_valle_nocturno'] = df.index.hour.isin([1, 2, 3, 4, 5, 6]).astype(int)
    
    # Pico alto real (horas 11, 19, 20: consumo >P75)
    df['es_pico_alto'] = df.index.hour.isin([11, 19, 20]).astype(int)
    
    # Rampa ascendente matutina (7-11)
    df['es_rampa_ascendente'] = df.index.hour.isin([7, 8, 9, 10]).astype(int)
    
    # Meseta alta sostenida (11-21)
    df['es_meseta_alta'] = ((df.index.hour >= 11) & (df.index.hour <= 21)).astype(int)
    
    # Descenso nocturno (22-24)
    df['es_descenso_nocturno'] = df.index.hour.isin([22, 23, 0]).astype(int)
    
    # === RATIOS  ===
    
    if 'Demanda' in df.columns:
        # Promedio del valle nocturno (últimas 6 horas del valle: 1-6 AM)
        df['promedio_valle_reciente'] = df['Demanda'].rolling(
            window=12, min_periods=6  # 6 horas = 12 períodos de 30min
        ).mean()
        
        # Promedio de la meseta alta (últimas 4 horas de meseta)
        df['promedio_meseta_reciente'] = df['Demanda'].rolling(
            window=8, min_periods=4  # 4 horas = 8 períodos de 30min
        ).mean()

        df = df.dropna()
    
    return df

# Ahora aplicarlo correctamente a tu dataframe
datos_fe_final = crear_features_demanda_especificos(datos_fe)

print(f"Features originales: {datos_fe.shape[1]}")
print(f"Features con nuevas variables: {datos_fe_final.shape[1]}")
print(f"Nuevas features agregadas: {datos_fe_final.shape[1] - datos_fe.shape[1]}")
```

```python
# === SEPARAR FEATURES Y TARGET ===
target = 'Demanda'
features_modelo = [
    # Temporales originales
    'ciclo_diario', 'dia_0Domingo', 'dia_1Lunes', 'dia_2Martes', 
    'dia_3Miércoles', 'dia_4Jueves', 'dia_5Viernes', 'dia_6Sábado', 'feriado',
    
    # Patrones específicos
    'es_valle_nocturno', 'es_pico_alto', 'es_rampa_ascendente', 
    'es_meseta_alta', 'es_descenso_nocturno',
    
]

X = datos_fe_final[features_modelo]
y = datos_fe_final[target]

print(f"Features shape: {X.shape}")
print(f"Target shape: {y.shape}")
```

```python
from sklearn.model_selection import train_test_split

# Para series temporales, NUNCA usar random split

# Opción A: División simple temporal
train_size = int(len(datos_fe_final) * 0.7)  # 70% train
val_size = int(len(datos_fe_final) * 0.15)   # 15% validation
# 15% restante para test

X_train = X.iloc[:train_size]
y_train = y.iloc[:train_size]

X_val = X.iloc[train_size:train_size+val_size]
y_val = y.iloc[train_size:train_size+val_size]

X_test = X.iloc[train_size+val_size:]
y_test = y.iloc[train_size+val_size:]

print("=== DIVISIÓN TEMPORAL ===")
print(f"Train: {X_train.shape[0]} observaciones ({X_train.index[0]} a {X_train.index[-1]})")
print(f"Val:   {X_val.shape[0]} observaciones ({X_val.index[0]} a {X_val.index[-1]})")
print(f"Test:  {X_test.shape[0]} observaciones ({X_test.index[0]} a {X_test.index[-1]})")
```

## **4.Modelado**

```python
datos_fe_final.head()
```

```python
#Se establece el horizonte de observación. En este caso una semana.
steps = 48*7
```

```python
#Se divide los datos en entrenamiento y prueba
x_train = datos_fe_final.drop(columns='Demanda')[:-steps]  # :-steps --> toma todo menos los ultimos registros segun la cantidad de registros de steps
y_train = datos_fe_final.Demanda[:-steps]
x_test = datos_fe_final.drop(columns='Demanda')[-steps:] # steps: --> toma solo los ultimos registros segun el valor de steps
y_test = datos_fe_final.Demanda[-steps:]
```

```python
#Se genera el modelo de forecasting recursivo
forecaster = ForecasterRecursive(
    regressor=XGBRegressor(random_state=50),
    lags= 48*7
)
```

Este modelo usará el algoritmo de XGBoost como motor de predicción y para predecir el valor actual el modelo considera la demanda de la última semana completa como variables de entrada (lags=48*7)

```python
#Se define la validación cruzada
cv = TimeSeriesFold(
      steps              = 48,
      initial_train_size = 12 * 7 * 48,
      refit              = False,
      fixed_train_size   = False
    )
```

* **Steps:** El tamaño de cada bloque de validación es de un día
* **initial_train_size:** La cantidad de observaciones del primer entrenamiento será de aproximadamente 3 meses.
* **texto en negrita**refit: Se indica que el modelo no se re-entrene en cada fold, esto para ahorrar tiempo de cómputo
* **fixed_train_size:** con *False* Se indica que el bloque de entrenamiento va creciendo a medida que avanza la validación

```python
#Entreno el modelo
forecaster.fit(y=y_train)
#Genero las predicciones para un horizonte de una semana
predictions = forecaster.predict(steps=steps)
```

```python
#Visualizo de forma gráfica
fig, ax = plt.subplots(figsize=(12,4))
y_train[-steps:].plot(ax=ax, label='Train')
y_test.plot(ax=ax, label='Test')
predictions.plot(ax=ax, label='predictions')
plt.title('Predicción de Modelo SKForecast - sin variables exógenas')
ax.legend()
plt.show()
```

***Métricas de error***

Se calcula el Mean Squared Error (MSE) y el Mean Absolute Error (MAE) comparando los valores reales con las predicciones. Mientras menor sean sus valores significa que mejor es la predicción.

```python
error_mse1 = mean_squared_error(y_test, predictions)
error_mae1 = mean_absolute_error(y_test, predictions)
print(f'MSE: {np.sqrt(error_mse1)}')
print(f'MAE: {error_mae1}')
```

**AGREGO LAS VARIABLES EXOGENAS**

```python
forecaster = ForecasterRecursive(
    regressor=XGBRegressor(max_depth=3, n_estimators=75, random_state=50),
    lags= [48, 48*2, 48*7] # el modelo considera la demanda exacta de hace un día, de dos dias atras, una semana atrás y un mes atrás
)
forecaster.fit(y=y_train, exog=x_train)
predictions = forecaster.predict(steps=steps, exog=x_test)
```

```python
fig, ax = plt.subplots(figsize=(12,4))
y_train[-steps:].plot(ax=ax, label='Train')
y_test.plot(ax=ax, label='Test')
predictions.plot(ax=ax, label='predictions')
plt.title('Predicción de Modelo SKForecast- con variables exógenas')
ax.legend()
plt.show()
```

***Métricas de error***

```python
error_mse1 = mean_squared_error(y_test, predictions)
error_mae1 = mean_absolute_error(y_test, predictions)
print(f'MSE: {np.sqrt(error_mse1)}')
print(f'MAE: {error_mae1}')
```

```python
# Análisis de Residuos
residuos = y_test - predictions
plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.scatter(predictions, residuos, alpha=0.5)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Predicciones')
plt.ylabel('Residuos')
plt.title('Residuos vs Predicciones')

plt.subplot(1, 3, 2)
residuos.hist(bins=30)
plt.title('Distribución de Residuos')

plt.subplot(1, 3, 3)
plt.plot(residuos.index, residuos)
plt.title('Residuos en el Tiempo')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
